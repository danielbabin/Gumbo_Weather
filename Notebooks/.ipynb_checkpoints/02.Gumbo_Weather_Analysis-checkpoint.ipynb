# Gumbo Weather Analysis
The purpose of this notebook is to compare historical temperature data from New Orleans, Louisiana to climate projections and assess the likelihood of sub-50 degree (F) days, optimal weather for Gumbo as identified by my grandfather, occurring during winter in different climatological intervals. We have 4 intervals. Two (between 1960-1990 and 1990-2020) come from historical data. Two intervals (2020-2050 and 2050-2080) use data from climate models running scenarios for [Relative Concentration Pathway 4.5](https://en.wikipedia.org/wiki/Representative_Concentration_Pathway), accessed in the previous notebook.  
<br>
In the first section, we **import, manage and restructure the climate records**. The historical record pieces together records from two weather stations, and those records need to be combined. The individual climate models need to be appraised individually, and then averaged into an ensemble. Both historical records and climate models need to be sliced into 30 year climatological intervals, and averaged/statistically analyzed across them.  
<br>
In the second section, we **compare those climatologies to a threshold to calculate the probability of any day of the year being a sub 50 degree day**. This can be done as simply as counting the number of days with temperatures below 50 degrees, and expressing it as a percentage of the total days in the interval.

## Climate Model Bias Correction
Each climate model operates with slightly different physics, and climate models are known to be subject to bias. This bias can be accounted for by comparing a historical output of the model to historical climate data in order to assess the sense in which variability produced by the climate is different from real climate variability.  

We use [quantile mapping](https://hess.copernicus.org/articles/16/3309/2012/) for our bias correction method. It is a good method for maintaining behavoir at statistical extremes, used by the [New York City Panel of Climate Change's Temperature Projections](https://www1.nyc.gov/site/orr/challenges/nyc-panel-on-climate-change.page).  

The method sorts the data in bins (40 in our case) which contain an equal number data points. You them calculate the average difference between, for example, the models lowest 2.5% of temperatures and the lowest 2.5% of temperatures from observations. We assume that this kind of bias in the model would be consistent through time, and apply this correction to future intervals of interest. Graphics below visualizes how this process works. 
  
The "historical" simulations for these climate models end at 2015, so we take that climatology (1985-2015) from the models, and compare it to the climatology from that same time interval in the historical data.

```python
obs_stats=pd.DataFrame(index=np.arange(1,41))
model_stats=pd.DataFrame(index=np.arange(1,41))
```

This code goes through each model, bins the data, and calculates the difference in the mean from each bin for each month.

```python
biases={}
for model in model_list:
    for i,month in enumerate(range(1,13)):
        obs=observations.loc[observations.index.month==month].to_frame()
        while obs.loc[obs['TMAX_STACK'].duplicated(),'TMAX_STACK'].any()==True:
                obs.loc[obs['TMAX_STACK'].duplicated(),'TMAX_STACK']=obs.loc[obs['TMAX_STACK'].duplicated(),
                                                                             'TMAX_STACK']+.001
        obs['bin']=pd.qcut(obs['TMAX_STACK'],40,labels=np.arange(1,41))
        obs_stats[month]=obs.groupby(by='bin').mean()['TMAX_STACK']

        model_output=((tables['historical'].loc[tables['historical'].index.month==month,
                                                model])).to_frame()
        model_output['bin']=pd.qcut(model_output[model],40,labels=np.arange(1,41))
        model_stats[month]=model_output.groupby(by='bin').mean()[model]
    biases[model]=obs_stats-model_stats
```

### Implement Bias Correction
You can see the results of applying the  correction for the mean based on our observation of the bias in the historical runs of the model. The result is a very good match in the shape of the variability. 

```python
model='GFDL-CM4'
```

```python
bias_corr_test=tables['historical'][model].copy()
for i,month in enumerate(range(1,13)):
    model_output=((tables['historical'].loc[tables['historical'].index.month==month,
                                                model])).to_frame()
    model_output['bin']=pd.qcut(model_output[model],40,labels=np.arange(1,41))
    for b in model_output['bin'].unique():
        ids=model_output.loc[model_output['bin']==b].index
        bias_corr_test.loc[ids]=model_output.loc[ids,model]+biases[model].loc[b,month]
```

```python
fig,axes=plt.subplots(nrows=12,ncols=3,figsize=(10,12),sharex=True)
for i,month in enumerate(range(1,13)):
    obs=observations.loc[observations.index.month==month].copy()
    obs.plot.hist(ax=axes[i,0],bins=np.linspace(20,110,91))
    axes[i,0].annotate(mon_list[i],xy=(.02,.95),xycoords='axes fraction',ha='left',va='top',size=12)
    
    model_output=(bias_corr_test.loc[bias_corr_test.index.month==month])
    model_output.plot.hist(ax=axes[i,1],bins=np.linspace(20,110,91),color='tab:orange')
    axes[i,1].annotate(mon_list[i],xy=(.02,.95),xycoords='axes fraction',ha='left',va='top',size=12)
    
    model_output.plot.hist(ax=axes[i,2],bins=np.linspace(20,110,91),color='tab:orange',alpha=.3)
    obs.plot.hist(ax=axes[i,2],bins=np.linspace(20,110,91),alpha=.3)
    
axes[0,0].set_title('Historical Observations (1985-2014)')
axes[0,1].set_title(model+' Output (1985-2014)')
axes[0,2].set_title('Both')

plt.tight_layout()
plt.subplots_adjust(hspace=.09)
```

```python
bias_corrected=tables['ssp245'].iloc[:,5:].copy()
```

The code below takes those biases and applies them to the model projections. 

```python
for model in bias_corrected.columns:
    if model in biases.keys():
        for i,month in enumerate(range(1,13)):
            model_output=(bias_corrected.loc[bias_corrected.index.month==month,
                                              model]).to_frame()
            model_output['bin']=pd.qcut(model_output[model],40,labels=np.arange(1,41))
            for b in model_output['bin'].unique():
                ids=model_output.loc[model_output['bin']==b].index
                bias_corrected.loc[ids,model]=model_output.loc[ids,model]+biases[model].loc[b,month]
    else:
        bias_corrected=bias_corrected.drop(model,axis=1)
```

### Gumbo Weather Statistics
This is the part where we just count the number of days below our temperature threshold, 50 degrees F. The analysis of gumbo weather can be presented in two ways. The number of gumbo days in a given month over a given interval, or the probability of a gumbo day occurring on a certain day in a given month. The second can be done by expressing the total number of gumbo days over the total number of days in the interval of interest (number years in the interval of interest * number of days in the month of interest)

```python
gumbo_days=pd.DataFrame(index=range(1,13))
```

```python
days_in_month=np.array([31,28,31,30,31,30,31,31,30,31,30,31])
```

#### Historical

```python
time_bounds=[['1960','1989'],['1990','2019']]
for slc in time_bounds:
    df=stack.loc[slc[0]:slc[1]].copy()
    for i,month in enumerate(range(1,13)):
        name=slc[0]+'-'+str(int(slc[0])+29)
        gumbo_days.loc[month,name+' Gumbo Days (n)']=(df.loc[df.index.month==month,'TMAX_STACK']<50).sum()
        gumbo_days.loc[month,name+' Probability (%)']=((df.loc[df.index.month==month,'TMAX_STACK']<50).sum()/
                                                  30/days_in_month[month-1]*100)
```

#### Projections

```python
projected_gumbo_days={}
projected_gumbo_days['2020']=pd.DataFrame(columns=bias_corrected.columns,index=range(1,13))
projected_gumbo_days['2050']=pd.DataFrame(columns=bias_corrected.columns,index=range(1,13))
projected_gumbo_days_ensemble=pd.DataFrame(index=range(1,13),columns=['2020','2050'])
```

```python
time_bounds=[['2020','2049'],['2050','2079']]
for slc in time_bounds:
    df=bias_corrected.loc[slc[0]:slc[1]].copy()
    for model in bias_corrected.columns:
        for i,month in enumerate(range(1,13)):
            projected_gumbo_days[slc[0]].loc[month,model]=(df.loc[df.index.month==month,model]<50).sum()
```

```python
gumbo_days['2020-2050 Gumbo Days (n)']=projected_gumbo_days['2020'].mean(axis=1).astype(int)
gumbo_days['2020-2050 Probability (%)']=gumbo_days['2020-2050 Gumbo Days (n)']/30/days_in_month*100
gumbo_days['2050-2079 Gumbo Days (n)']=projected_gumbo_days['2050'].mean(axis=1).astype(int)
gumbo_days['2050-2079 Probability (%)']=gumbo_days['2050-2079 Gumbo Days (n)']/30/days_in_month*100
```

```python
gumbo_days['Month']=mon_list
gumbo_days=gumbo_days.set_index('Month')
gumbo_days=gumbo_days.loc[mon_list[7:]+mon_list[:7]]
gumbo_days=gumbo_days.round(1)
```
