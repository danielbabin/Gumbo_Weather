# Basemap

```python
from cartopy.io.shapereader import Reader
from cartopy.feature import ShapelyFeature

shp_path='/Users/danielbabin/Documents/LamontColumbia/U1477/Data/Shapefiles/'
fname = 'ne_50m_rivers_lake_centerlines_scale_rank.shp'
fname2 = 'ne_50m_lakes.shp'

sf_riv = ShapelyFeature(Reader(shp_path+fname).geometries(),
                                ccrs.PlateCarree(), facecolor='none')
sf_lks = ShapelyFeature(Reader(shp_path+fname2).geometries(),
                                ccrs.PlateCarree(), facecolor='none')
states_provinces = cfeature.NaturalEarthFeature(
        category='cultural',
        name='admin_1_states_provinces_lines',
        scale='50m',
        facecolor='none')
```

```python
def plotBase():
    top=33.5
    bot=28.5
    left=-96
    right=left+(top-bot)/9*18.65

    ax=fig.add_axes([0,0,1,1],projection=ccrs.Mercator.GOOGLE)
    ax.set_extent([left,right,bot,top])

    ax.add_feature(cfeature.OCEAN,zorder=1,color='#EAF5FA',alpha=.3)
    ax.add_feature(cfeature.COASTLINE,zorder=2)
    ax.add_feature(cfeature.LAND,zorder=0,alpha=.5)
    ax.add_feature(sf_lks,facecolor='#EAF5FA',edgecolor='k',alpha=.3)
    ax.add_feature(states_provinces, edgecolor='k')
    return ax
    
fig=plt.figure(figsize=(16,9),facecolor='w')
plotBase()

fig.savefig(scrap_path+'test_map.png')
```

# Add Vegetation Data

```python
def makeVegetationMap(collection,slicename,vmin=.2,vmax=.8):
    ax=plotBase()
    cax=collection[str(slicename)].sel(band=1).plot.imshow(robust=True,ax=ax,cmap='viridis',vmin=vmin,vmax=vmax,
                                                           add_colorbar=False)
    ax.set_title('')

    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    axins = inset_axes(ax, width="100%",height="100%",loc=3,bbox_transform=ax.transAxes,
                        bbox_to_anchor=(.02, .03, .03, .94),borderpad=0)
    cbar=fig.colorbar(cax,cax=axins)
    
    xpos1=.115
    ax.annotate('                                                 ',xy=(xpos1,.5),xycoords='figure fraction',
                fontsize=14,rotation=90,ha='center',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    cbar.ax.tick_params(labelsize=14) 
    
    center_of_words=.23
    ax.annotate('More Green Vegetation',xy=(xpos1+.001,center_of_words),xycoords='figure fraction',
                fontsize=14,rotation=90,ha='center',va='bottom',fontname='serif')
    ax.annotate('',xy=(xpos1,center_of_words+.53),xytext=(xpos1,center_of_words+.43),
                xycoords='figure fraction',arrowprops=dict(facecolor='black'))
    
    ### Annotations
    top = .58
    xpos=.97
    ax.annotate('Has November gotten greener?',xy=(xpos,top),xycoords='figure fraction',
                fontname='serif',fontsize=20,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    ax.annotate('Nov 1st, '+str(slicename),xy=(xpos,top-.12),xycoords='figure fraction',
                fontname='serif',fontsize=30,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
```

## Calculate Anomalies

```python
def makeAnomalyMap(year):
    ax=plotBase()
    baseline=(maskedFallNDVIs['2000'].sel(band=1)+
              maskedFallNDVIs['2001'].sel(band=1)+
              maskedFallNDVIs['2002'].sel(band=1))/3
    
    cax=(maskedFallNDVIs[str(year)].sel(band=1)-baseline).plot.imshow(robust=True,ax=ax,cmap='coolwarm',
                                                                  vmin=-.4,vmax=.4,add_colorbar=False)
    ax.set_title('')

    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    axins = inset_axes(ax, width="100%",height="100%",loc=3,bbox_transform=ax.transAxes,
                        bbox_to_anchor=(.02, .03, .03, .94),borderpad=0)
    cbar=fig.colorbar(cax,cax=axins)
    
    xpos_label=.145
    ax.annotate('                                                                              ',
                xy=(xpos_label,.5),xycoords='figure fraction',
                fontsize=14,rotation=90,ha='center',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    cbar.ax.tick_params(labelsize=14) 
    
    ax.annotate("Greener than '00-'02",xy=(xpos_label+.001,.93),
                xycoords='figure fraction',fontsize=14,rotation=90,ha='center',va='top')
    ax.annotate("Deader than '00-'02",xy=(xpos_label+.001,.08),
                xycoords='figure fraction',fontsize=14,rotation=90,ha='center',va='bottom')
    
    ax.annotate('',xy=(xpos_label,.56),xytext=(xpos_label,.43),
                xycoords='figure fraction',arrowprops=dict(arrowstyle='<|-|>', lw=3))
    
    ### Annotations
    top = .58
    xpos=.97
    ax.annotate('Has November gotten greener?',xy=(xpos,top),xycoords='figure fraction',
                fontname='serif',fontsize=20,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    ax.annotate('Nov 1st, '+str(year),xy=(xpos,top-.12),xycoords='figure fraction',
                fontname='serif',fontsize=30,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
```

```python
weeklyNDVIs={}
maskedWeeklyNDVIs={}
for wk in weekly_filenames:
    weeklyNDVIs[wk[:-9]]=rxr.open_rasterio(weekly_data_path+wk,masked=True)
    maskedWeeklyNDVIs[wk[:-9]]=weeklyNDVIs[wk[:-9]].where(weeklyNDVIs[wk[:-9]]!=0)
```

```python
def getMonth(n_week):
    if (n_week<=3)|(n_week==52):    #4
        return 'February'
    elif (n_week>=4)&(n_week<=7):   #4
        return 'March'
    elif (n_week>=8)&(n_week<=12):  #5
        return 'April'
    elif (n_week>=13)&(n_week<=16): #4
        return 'May'
    elif (n_week>=17)&(n_week<=20): #4
        return 'June'
    elif (n_week>=21)&(n_week<=25): #5
        return 'July'
    elif (n_week>=26)&(n_week<=29): #4
        return 'August'
    elif (n_week>=30)&(n_week<=33): #4
        return 'September'
    elif (n_week>=34)&(n_week<=38): #5
        return 'October'
    elif (n_week>=39)&(n_week<=42): #4
        return 'November'
    elif (n_week>=43)&(n_week<=46): #4
        return 'December'
    elif (n_week>=47)&(n_week<=51): #4
        return 'January'
```

```python
def makeAnnualVegetationMap(collection,slicename,vmin=.2,vmax=.8):
    ax=plotBase()
    cax=collection[str(slicename)].sel(band=1).plot.imshow(robust=True,ax=ax,cmap='viridis',vmin=vmin,vmax=vmax,
                                                           add_colorbar=False)
    ax.set_title('')

    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    axins = inset_axes(ax, width="100%",height="100%",loc=3,bbox_transform=ax.transAxes,
                        bbox_to_anchor=(.02, .03, .03, .94),borderpad=0)
    cbar=fig.colorbar(cax,cax=axins)
    
    xpos1=.115
    ax.annotate('                                                 ',xy=(xpos1,.5),xycoords='figure fraction',
                fontsize=14,rotation=90,ha='center',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    cbar.ax.tick_params(labelsize=14) 
    
    center_of_words=.23
    ax.annotate('More Green Vegetation',xy=(xpos1+.001,center_of_words),xycoords='figure fraction',
                fontsize=14,rotation=90,ha='center',va='bottom',fontname='serif')
    ax.annotate('',xy=(xpos1,center_of_words+.53),xytext=(xpos1,center_of_words+.43),
                xycoords='figure fraction',arrowprops=dict(facecolor='black'))
    
    ### Annotations
    top = .08
    xpos=.97
    ax.annotate("Seasonal Foliage Variability in Louisiana",xy=(xpos,top),xycoords='figure fraction',
                fontname='serif',fontsize=20,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
    month=getMonth(slicename)
    ax.annotate(month,xy=(xpos,top+.12),xycoords='figure fraction',
                fontname='serif',fontsize=30,fontweight='bold',ha='right',va='center',
                bbox=dict(boxstyle='round',fc="w", ec="k",))
```
